# Test of Package 

This notebook is used to show that the package works.

```python
from pygbm.gbm_simulation import GBMSimulator 
import matplotlib.pyplot as plt 

# Parameters for GBM
y0 = 1.0 
mu = 0.05
sigma = 0.2 
T = 1.0 
N = 100

# Initialize Simulator 
simulator = GBMSimulator(y0, mu, sigma)

#Simulate Path 
t_values, y_values = simulator.simulate_path(T, N)

# Plot Simulated Path 
plt.plot(t_values, y_values, label = "GMB Path")
plt.xlabel("Time")
plt.ylabel("Y(t)")
plt.title("Simulated Geometric Brownian Motion")
plt.legend()
plt.show()
```
